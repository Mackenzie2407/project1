```python
cleaned_h.isnull().sum()
```

```python
cleaned_h.Year = cleaned_h.Year.astype('int')
cleaned_h = cleaned_h.reset_index()
```

```python
cleaned_h["EU_Sales"] = cleaned_h["EU_Sales"].str.replace("M", "")
cleaned_h["EU_Sales"] = cleaned_h["EU_Sales"].str.replace("m", "")
cleaned_h["JP_Sales"] = cleaned_h["JP_Sales"].str.replace("M", "")
cleaned_h["JP_Sales"] = cleaned_h["JP_Sales"].str.replace("m", "")
cleaned_h["NA_Sales"] = cleaned_h["NA_Sales"].str.replace("M", "")
cleaned_h["NA_Sales"] = cleaned_h["NA_Sales"].str.replace("m", "")
cleaned_h["Other_Sales"] = cleaned_h["Other_Sales"].str.replace("M", "")
cleaned_h["Other_Sales"] = cleaned_h["Other_Sales"].str.replace("m", "")
cleaned_h.head(3)
```

```python
error_jp = cleaned_h[cleaned_h['JP_Sales'].str.contains('K')]
error_jp['JP_Sales'] = error_jp.JP_Sales.str.replace('K', '', regex = True)
error_jp['JP_Sales'] = error_jp['JP_Sales'].astype('float')
error_jp['JP_Sales'] = error_jp['JP_Sales']/1000
error_jp
```

```python
error_jp1 = error_jp[['Unnamed: 0', 'JP_Sales']]
error_jp1
```

```python
error_eu = cleaned_h[cleaned_h['EU_Sales'].str.contains('K')]
error_eu['EU_Sales'] = error_eu.EU_Sales.str.replace('K', '', regex = True)
error_eu['EU_Sales'] = error_eu['EU_Sales'].astype('float')
error_eu['EU_Sales'] = error_eu['EU_Sales']/1000
```

```python
error_eu1 = error_eu[['Unnamed: 0', 'EU_Sales']]
error_eu1
```

```python
error_na = cleaned_h[cleaned_h['NA_Sales'].str.contains('K')]
error_na['NA_Sales'] = error_na.NA_Sales.str.replace('K', '', regex = True)
error_na['NA_Sales'] = error_na['NA_Sales'].astype('float')
error_na['NA_Sales'] = error_na['NA_Sales']/1000
```

```python
error_na1 = error_na[['Unnamed: 0', 'NA_Sales']]
error_na1
```

```python
error_other = cleaned_h[cleaned_h['Other_Sales'].str.contains('K')]
error_other['Other_Sales'] = error_other.Other_Sales.str.replace('K', '', regex = True)
error_other['Other_Sales'] = error_other['Other_Sales'].astype('float')
error_other['Other_Sales'] = error_other['Other_Sales']/1000
```

```python
error_other1 = error_other[['Unnamed: 0', 'Other_Sales']]
error_other1
```

```python
cleaned_h=error_jp1.set_index('Unnamed: 0').combine_first(cleaned_h.set_index('Unnamed: 0')).reset_index()
cleaned_h=error_eu1.set_index('Unnamed: 0').combine_first(cleaned_h.set_index('Unnamed: 0')).reset_index()
cleaned_h=error_na1.set_index('Unnamed: 0').combine_first(cleaned_h.set_index('Unnamed: 0')).reset_index()
cleaned_h=error_other1.set_index('Unnamed: 0').combine_first(cleaned_h.set_index('Unnamed: 0')).reset_index()
cleaned_h
```
